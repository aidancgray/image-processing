```python
import os
import glob
import math
import numpy as np
from astropy.io import fits
```

```python
def open_fits(filename):
    fitsFile = fits.open(filename)
    fitsData = fitsFile[0].data
    fitsHdr = fitsFile[0].header

    return fitsData, fitsHdr
```

```python
os.chdir('/Users/aidancgray/Documents/SCORPIO/SVC/Images/08-11-2023/good/dsi/post/')
directoryList = glob.glob('*.fits')
# directoryList.sort(key=lambda f: int(''.join(filter(str.isdigit, f))))
directoryList.sort()

results = {}
pxScale = 4.539062  # microns

for file in directoryList:
    dd,  hdr  = open_fits(file)
    yUp = dd.shape[0]
    xUp = dd.shape[1]

    total_flux = 0.0
    x_mean = 0.0
    y_mean = 0.0
    x_m2 = 0.0
    y_m2 = 0.0

    for row in range(yUp):
        for col in range(xUp):
            pxl_flux = dd[row, col]
            if pxl_flux > 0.0:
                tmp = total_flux + pxl_flux
                x = (col + 0.5) * pxScale
                y = (row + 0.5) * pxScale
                delta_x = x - x_mean
                delta_y = y - y_mean
                Rx = delta_x * pxl_flux / tmp
                Ry = delta_y * pxl_flux / tmp
                x_mean = x_mean + Rx
                y_mean = y_mean + Ry
                x_m2 = x_m2 + (total_flux * delta_x * Rx)
                y_m2 = y_m2 + (total_flux * delta_y * Ry)
                total_flux = tmp

    xvar = x_m2 / total_flux
    yvar = y_m2 / total_flux
    x_sigma = math.sqrt(xvar)
    y_sigma = math.sqrt(yvar)
    rms = math.sqrt((x_sigma * x_sigma) + (y_sigma * y_sigma))
    # print(f'rms={rms}')
    results.update({file: round(rms, 3)})
```

```python
for file in results:
    print(file,'->',results[file])
    #print(results[file])
for file in results:
    #print(file,'->',results[file])
    print(results[file])
```
